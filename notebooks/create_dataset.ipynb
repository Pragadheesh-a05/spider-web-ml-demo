```python
import os
import pandas as pd
import numpy as np
from PIL import Image

# Create dataset folder
base_path = "../dataset"
classes = ["clean_web", "damaged_web", "no_web"]

for cls in classes:
    os.makedirs(os.path.join(base_path, cls), exist_ok=True)

# Generate dummy images
data = []

for cls in classes:
    for i in range(20):
        img = np.random.randint(0, 255, (128,128,3), dtype=np.uint8)
        img_path = f"{base_path}/{cls}/{cls}_{i}.png"
        Image.fromarray(img).save(img_path)

        data.append({
            "image_name": f"{cls}_{i}.png",
            "label": cls
        })

# Save Excel dataset
df = pd.DataFrame(data)
excel_path = "../dataset/spider_web_dataset.xlsx"
df.to_excel(excel_path, index=False)

print("Dataset + Excel created successfully!")
```
